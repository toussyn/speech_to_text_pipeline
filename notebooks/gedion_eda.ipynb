```python
data.groupby('category').count().sort_values(by='views',ascending=False)
```

ሀገር አቀፍ ዜና meaning Domestic News is the largest category followed by ስፖርት meaing Sport is the second largest catefory and መዝናኛ meaning Entertainment has the least amount of articles.

```python
data.groupby('category')['views'].sum()
```

ቢዝነስ meaning Business has the highest number of views while ዓለም አቀፍ ዜና meaning International News has the lowest number of views.

```python
data[data['views']=='Unknown'].groupby('category').count()
```
